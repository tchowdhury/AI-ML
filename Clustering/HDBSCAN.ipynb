# HDBSCAN Clustering

```python
# HDBSCAN Custering
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset

dataset = pd.read_csv('Mall_customers.csv')
data = dataset.iloc[:,[3,4]].values
data.shape
```

```python
# visualizing dataset

plt.scatter(data[:,0], data[:,1],s = 10, c = 'red')
```

```python
# Fitting the HDBSCAN to dataset
import hdbscan

hdbscan = hdbscan.HDBSCAN(min_cluster_size = 10, min_samples= 5)
labels = hdbscan.fit_predict(data)
hdbscan.condensed_tree_.plot(select_clusters= True)
np.unique(labels)

# hdbscan.single_linkage_tree_.plot()
```

```python
# Visualizing the clusters
plt.scatter(data[labels == -1,0], data[labels == -1,1], s = 10, c = 'orange')
plt.scatter(data[labels == 0,0], data[labels == 0,1], s = 10, c = 'blue')
plt.scatter(data[labels == 1,0], data[labels == 1,1], s = 10, c = 'red')
plt.scatter(data[labels == 2,0], data[labels == 2,1], s = 10, c = 'pink')
plt.scatter(data[labels == 3,0], data[labels == 3,1], s = 10, c = 'green')
plt.scatter(data[labels == 4,0], data[labels == 4,1], s = 10, c = 'brown')

# draw the centriod
plt.title('Clusters result on the dataset')
plt.xlabel('Annual Income')
plt.ylabel('Spemding Score')
plt.show()
```

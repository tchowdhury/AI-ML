# KMeans Limitation with Different Cluster Sizes

```python
# Kmeans Custering
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing dataset

dataset = pd.read_csv('Aggregation.csv')
data = dataset.iloc[:,[0,1]].values
data.shape
```

```python
# visualizing dataset

plt.scatter(data[:,0], data[:,1],s = 10, c = 'black')
```

```python
# Elbow method for determining the number of cluster
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting the kmeans to dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10)
labels = kmeans.fit_predict(data)
np.unique(labels)
```

```python
# Visualizing the clusters
plt.scatter(data[labels == 0,0], data[labels == 0,1], s = 10, c = 'blue')
plt.scatter(data[labels == 1,0], data[labels == 1,1], s = 10, c = 'red')
plt.scatter(data[labels == 2,0], data[labels == 2,1], s = 10, c = 'pink')
plt.scatter(data[labels == 3,0], data[labels == 3,1], s = 10, c = 'green')
plt.scatter(data[labels == 4,0], data[labels == 4,1], s = 10, c = 'brown')

# draw the centriod
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 18, c = 'black')
plt.title('Clusters result on the dataset')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
